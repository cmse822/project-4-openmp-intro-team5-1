```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
matrix_sizes = np.array([1, 10, 25, 50, 80, 100, 150, 200, 300, 400, 600, 1000, 1200, 1500, 2000, 3000, 4000])

omp = np.array([
    1.41151e-05, 1.53407e-05, 0.00020422, 0.00158539, 0.00645487, 
    0.0119882, 0.0412143, 0.0986895, 0.33485, 0.871513, 
    3.14439, 14.9014, 26.0034, 55.4776, 150.294, 535.798, 1362.04])

noomp = np.array([
    0, 1.3113e-05, 0.00019002, 0.00146103, 0.00592589, 
    0.011704, 0.0382981, 0.091434, 0.316908, 0.818844, 
    3.00286, 14.4603, 24.6308, 48.2817, 123.961, 513.444, 1334.11])

plt.figure(figsize=(10, 6))
plt.plot(matrix_sizes, omp, label="With OMP")
plt.plot(matrix_sizes, omp, label="Without OMP")
plt.title('Comparison of Performance With and Without OMP')
plt.xlabel('Matrix Size')
plt.ylabel('Time/s')
plt.legend()
plt.show()
```

```python
matrix_sizes = np.array([20, 100, 1000])
threads = np.array([1, 2, 3, 4, 5])
time_taken = np.array([
    [0.000108734, 0.0124473, 15.234],    
    [0.000199791, 0.00901864, 8.3126], 
    [0.000337396, 0.00697817, 5.86437],   
    [0.000196386, 0.00333114, 4.62509],  
    [0.000206448, 0.00316833, 3.1084]     
])

plt.figure(figsize=(10, 6))
for i, thread in enumerate(threads):
    plt.plot(matrix_sizes, time_taken[i], '-o', label=f'{thread} Threads')

plt.title('Performance of Number of Threads on Time Taken vs Matrix Size')
plt.xlabel('Matrix Size')
plt.ylabel('Time/s')
plt.legend()
plt.show()
```
